```python
import random

#Function for extracting numbers from a string i.e getting rid of special characters
def number_extraction(num):


    stripped = ''
    
    #if element is a digit, add it to a string else ignore it
    for i in range(len(num)):
        if num[i].isdigit():
            stripped += num[i]
    
    #need to use the stripped string as an integer so returning as int
    if stripped:
        return int(stripped)
    else:
        return None #return none if string is empty to handle edge case


rand = random.randint(1,100)#random number generataion

print("Random Num: " + str(rand))

guessed = False

guesses_left = 5

while(guesses_left != 0 and guessed == False):
    
    guessed_num = input("Guess a random number from 1 to 100: ")
    guessed_num = number_extraction(guessed_num)

    if guessed_num>0 and guessed_num <= 100:

        #IF statements for rules of the game
        if(guessed_num == rand):

            print("Correct! Congratulations.")
            guessed = True

        elif (guessed_num > rand):

            print("Too High")
            guesses_left = guesses_left - 1
            print("Guesses Left: " + str(guesses_left))

        else:

            print("Too Low")
            guesses_left = guesses_left - 1
            print("Guesses Left: " + str(guesses_left))

#if guesses all used up, end game. Guesses are not used up for invalid inputs
if guessed_num <= 0:
    print("You Lose. Correct Number was: " + str(rand))
```

```python
special_characters = [
    '!', '"', '#', '$', '%', '&', "'", '(', ')', '*', '+', ',', '-', '.', '/',
    ':', ';', '<', '=', '>', '?', '@', '[', '\\', ']', '^', '_', '`', '{', '|',
    '}', '~'
] #List of special characters found on a standard keyboard

stop_words = [
    'a' , 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'can', 'do', 'for', 'from', 'has', 'have', 'he',
    'in', 'is', 'it', 'just', 'my', 'of', 'on', 'or', 'that', 'the', 'their', 'they', 'to', 'was', 'we',
    'with', 'you', 'is', 'im' , 'my' , 'im'
] #List of common stop words

#function using string slicing and checks each element one at a time as special characters are all one character long
def removePunctuationMarks(inString):
    
    punc_removed = ""
    for i in range(0, len(inString)):
        if inString[i] in special_characters:
            continue
        else:
            punc_removed+=inString[i]
    
    if inString:
        return punc_removed
    else:
        return None
    
#function uses string slicing but checks entire words using space as a delimiter as stop words are not single characters
def removeStopWords(inString):

    stop_removed = ""
    temp = ""
    i = 0

    while i < len(inString):#check entire string

        if inString[i] == ' ':
            if temp and temp.lower() not in stop_words:#if temp not empty and temp not in stop word list, add it to the stripped list "stop removed"


                if stop_removed:
                    #adding a space if the list is non empty i.e not the first word
                    stop_removed += ' '

                stop_removed += temp
            #resetting the temp string to prep for next word
            temp = ""

        else:
            #this adds the character of the words currently being processed into the temp until space reached
            temp += inString[i]

        i += 1
    
    #edge case for the last word as last word will not have a space at the end so it cant be used a delimiter
    if temp and temp.lower() not in stop_words:

        if stop_removed:

            stop_removed += ' '

        stop_removed += temp
    
    return stop_removed
    
    
#output formatting
inStr = input("Enter a statement to check: ")
inStr = removePunctuationMarks(inStr)
inStr = removeStopWords(inStr)
print("Statement after stripping: " + inStr)
```

```python
import random


class Player:
    def __init__(self, name="Player"):
        self.name = name
        self.score = 0

    def getName(self):
        return self.name

    def getScore(self):
        return self.score

    def incScore(self, score):
        self.score += score

    def setScore(self, score):
        self.score = score


class DungeonGrid:

    def __init__(self):
        self.grid = [['E', 'E', 'E', 'E', 'E'],
                     ['E', 'E', 'E', 'E', 'E'],
                     ['E', 'E', 'E', 'E', 'E'],
                     ['E', 'E', 'E', 'E', 'E'],
                     ['E', 'E', 'E', 'E', 'E']]

    def initializeGrid(self):

        for i in range(5):  # Place 3 health potions

            x, y = random.randint(0, 4), random.randint(0, 4)

            while self.grid[x][y] != 'E':  #This makes sure overwriting is not done

                x, y = random.randint(0, 4), random.randint(0, 4)

            self.grid[x][y] = 'H'

        for i in range(5):  # Place 3 traps

            x, y = random.randint(0, 4), random.randint(0, 4)

            while self.grid[x][y] != 'E':  #This makes sure overwriting is not done

                x, y = random.randint(0, 4), random.randint(0, 4)

            self.grid[x][y] = 'T'

        return self.grid
    
    def printGrid(self):

        for row in self.grid:

            print(' '.join(row))  #prints the grid in an actual grid format instead of as a list 

            


class DungeonEscape(DungeonGrid, Player):#using inheritance as grid and player as a part of the game

    def __init__(self, name="Player"):

        Player.__init__(self, name)  #initialize Player attributes
        DungeonGrid.__init__(self)   #initialize DungeonGrid and set up the grid
        self.health = 100
        self.position = (0, 0)  #start at top-left corner
        self.grid = self.initializeGrid()  #initialize the dungeon grid

    def updateHealth(self, hp):#method to change health    

        self.health += hp

    def move_player(self, direction):#method to shift position
        row, col = self.position

        #basic edge case checks
        if (direction.upper() == "UP" and row > 0):

            row -= 1

        elif (direction.upper() == "DOWN" and row < 4):

            row += 1

        elif (direction.upper() == "RIGHT" and col < 4):

            col += 1

        elif (direction.upper() == "LEFT" and col > 0):
            col -= 1

        else:
            print("Cannot move beyond grid boundary")

        self.position = (row, col)
        return self.position

    def check_cell(self, position):
        #checking position player just moved to
        row, col = position
        if (self.grid[row][col] == "T"):

            self.updateHealth(-20)
            print("Trap Triggered")

        elif (self.grid[row][col] == "H"):

            self.updateHealth(10)
            self.incScore(10)
            print("Health Potion Found")

        else:

            print("Empty Cell.")

        self.grid[row][col] = 'E'

    def play_turn(self):

        if self.health <= 0:

            print("Health depleted. Womp Womp Womp")
            return False
        elif self.score >= 49:
            print("Max Score Reached. You win")
            return False
        else:

            direction = input("Enter a direction you want to move in: ")

            directionList = ("UP","DOWN","LEFT","RIGHT")
            if (direction.upper() not in directionList):

                print("Invalid Move. Choose from Up, Down, Left or Right")

            else:

                self.move_player(direction.upper())
                self.check_cell(self.position)

                return True
            

    def getHealth(self):
        return self.health


game = DungeonEscape("Player")
game.printGrid()

while game.play_turn() == True:

    game.printGrid()
    print(game.getScore())
    print(game.getHealth())


    continue
```
