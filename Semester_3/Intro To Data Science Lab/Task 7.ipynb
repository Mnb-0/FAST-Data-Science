```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Task 1

```python
data = pd.read_csv('batsman_season_record.csv')
kohli = data[data['batsman'] == 'V Kohli'].iloc[0, 1:].values
sharma = data[data['batsman'] == 'RG Sharma'].iloc[0, 1:].values
seasons = data.columns[1:]

plt.plot(seasons, kohli, label='Virat Kohli', linestyle='-', marker='o', color='blue')
plt.plot(seasons, sharma, label='Rohit Sharma', linestyle='--', marker='s', color='green')

plt.axhline(y=kohli.mean(), color='blue', linestyle=':', label='Kohli Avg')
plt.axhline(y=sharma.mean(), color='green', linestyle='-.', label='Sharma Avg')

plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title('Performance Comparison: Virat Kohli vs. Rohit Sharma')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

Task 2

```python

batter = pd.read_csv('batter.csv')
batter = batter.head(50)

avg = batter['avg']
sr = batter['strike_rate']
players = batter['batter']

mask = np.isfinite(avg) & np.isfinite(sr)
avg_clean = avg[mask]
sr_clean = sr[mask]
players_clean = players[mask]

unique_players = players_clean.unique()
colors = plt.cm.get_cmap('tab10', len(unique_players))  # Color map

for i, player in enumerate(unique_players):
    player_mask = players_clean == player
    plt.scatter(avg_clean[player_mask], sr_clean[player_mask],
                color=colors(i / len(unique_players)), label=player)  # Normalize i

coefficients = np.polyfit(avg_clean, sr_clean, 1)
trendline = np.polyval(coefficients, avg_clean)

plt.plot(avg_clean, trendline, color='red', linestyle='--', label='Trendline')

plt.title("Avg vs Strike Rate For Top 50")
plt.xlabel("Average")
plt.ylabel("Strike Rate")
plt.grid(True)
plt.legend()

plt.show()
```

Task 3

```python
data = pd.read_csv('batsman_season_record.csv')


batsmen = data['batsman'] 
years = ['2015', '2016', '2017'] 
runs_2015 = data['2015']
runs_2016 = data['2016']
runs_2017 = data['2017']


n_batsmen = len(batsmen) 
index = np.arange(n_batsmen)  # The x locations for the bars
bar_width = 0.2

plt.bar(index, runs_2015, bar_width, label='2015', color='b')
plt.bar(index + bar_width, runs_2016, bar_width, label='2016', color='g')
plt.bar(index + 2 * bar_width, runs_2017, bar_width, label='2017', color='r')

plt.xlabel('Batsman')
plt.ylabel('Runs Scored')
plt.title('Batsmen Performance Across Seasons')
plt.xticks(index + bar_width, batsmen, rotation=45) 
plt.legend()  

plt.grid(True, axis='y', linestyle='--', alpha=0.7) 
plt.tight_layout()

plt.show()
```
