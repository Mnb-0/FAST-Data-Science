```python
import requests
from bs4 import BeautifulSoup
import csv

def fetch_and_parse(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        return soup
    else:
        print(f"Failed to fetch the page. Status code: {response.status_code}")
        return None

def extract_articles(soup):
    articles = []

    # Find all article elements with class "articles-list_item"
    article_elements = soup.find_all("div", class_="articles-list_item")
    print(f"Found {len(article_elements)} articles.")  # Debugging

    for article in article_elements:
        # Extract title
        title_tag = article.find("div", class_="head").find("a")
        title = title_tag.text.strip() if title_tag else "No Title"
        link = title_tag["href"] if title_tag else "No Link"

        # Extract description
        desc_tag = article.find("div", class_="text")
        description = desc_tag.text.strip() if desc_tag else "No Description"

        # Extract tags
        tags = []
        tag_elements = article.find("div", class_="tags").find_all("a")
        if tag_elements:
            tags = [tag.text.strip() for tag in tag_elements]
        tags = ", ".join(tags) if tags else "No Tags"

        # Append the data
        articles.append({
            "Title": title,
            "Link": link,
            "Description": description,
            "Tags": tags
        })

    return articles

def save_to_csv(articles, filename):
    # Define CSV headers
    headers = ["Title", "Link", "Description", "Tags"]

    # Open CSV file for writing
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=headers)

        # Write header to CSV
        writer.writeheader()

        # Write articles data to CSV
        for article in articles:
            writer.writerow(article)

    print(f"Data saved to {filename}")

# Main logic
url = "https://www.geeksforgeeks.org/basic/"
soup = fetch_and_parse(url)

if soup:
    articles = extract_articles(soup)
    if articles:
        save_to_csv(articles, "articles.csv")
    else:
        print("No articles found.")
```
